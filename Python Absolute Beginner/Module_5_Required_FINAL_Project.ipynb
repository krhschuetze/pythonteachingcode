### Additional Details
  
- initialize `total` variable which will sum integer values entered  
- initialize `items` variable which will build a string of the integer inputs separated with a *new line character*  
- define the `adding_report` function with one parameter `report` that will be a string with default of "T"  
- inside the function build a forever loop (infinite while loop) and inside the loop complete the following  
  - use a variable to gather input (integer or "Q")  
  - check if the input string is a digit (integer) and if it is...  
    - add input iteger to total  
    - if report type is "A" add the numeric character(s) to the item string seperated by a new line  
  - if not a digit, check if the input string is "Q" or starts with a "Q", **if "Q"** then...  
    - if the report type is "A" print out all the integer items entered and the sum total  
    - if report type is "T" then print out the sum total only  
    - `break` out of while loop to end the function after printing the report ("A" or "T") 
  - if not a digit and if not a "Q" then print a message that the "input is invalid"

````python
# [ ] create, call and test the adding_report() function

def adding_report(userChoice):
    if userChoice == "A": 
        aList = []
        # Take in input
        print("Input an integer to add to the total or \"Q\" to quit: ")
        ainitialInput = input()
        if (ainitialInput.upper() != 'Q' and ainitialInput.upper() != "QUIT" ):
            if (ainitialInput.isnumeric()): aList.append(int(ainitialInput))
            else: print(ainitialInput, "is invalid input")
        aAdd = ainitialInput
        while(aAdd.upper() != "Q" and aAdd.upper() != "QUIT" ):
            print("Enter an integer or \"Q\"): ")
            aAdd = input()
            if (aAdd.upper() == 'Q'): break
            if (aAdd.isnumeric()): aList.append(int(aAdd))
            else: print(aAdd, "is invalid input")
        # Calculate output
        aTotal = 0
        print("\nItems")
        for aNumber in aList: 
            print(aNumber)
            aTotal += aNumber
        print("\nTotal")
        print(aTotal)
        print("Calculated by: KRHS\n")
    elif userChoice == "T": 
        tList = []
        # Take in input
        print("Input an integer to add to the total or \"Q\" to quit: ")
        tinitialInput = input()
        if (tinitialInput.upper() != 'Q' and tinitialInput.upper() != "QUIT" ):
            if (tinitialInput.isnumeric()): tList.append(int(tinitialInput))
            else: print(tinitialInput, "is invalid input")
        tAdd = tinitialInput
        while(tAdd.upper() != "Q" and tAdd.upper() != "QUIT" ):
            print("Enter an integer or \"Q\"): ")
            tAdd = input()
            if (tAdd.upper() == 'Q'): break
            if (tAdd.isnumeric()): tList.append(int(tAdd))
            else: print(tAdd, "is invalid input")
        # Calculate output
        tTotal = 0
        for tNumber in tList: 
            tTotal += tNumber
        print("\nTotal")
        print(tTotal)
        print("Calculated by: KRHS\n")

print("adding_report(A) demo")
adding_report("A")
print("adding_report(T) demo")
adding_report("T")
````
